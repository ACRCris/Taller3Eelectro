```python
import numpy as np
import matplotlib.pyplot as plt


theta = np.linspace(0.01, 4*np.pi, 10000)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'},figsize=(10,10))
ax.plot(theta, (np.sin((np.pi)*np.cos(theta))**2)/(np.sin(theta)**2), linewidth=2)
ax.plot(theta, (np.sin(theta)**2), linewidth=2)
ax.set_rmax(4)
plt.thetagrids([
    0,30,60,90,120,150,180,210,240,270,300,330, 360
])
ax.set_rlabel_position(22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("Diagrama de radiación")

fig.savefig("polar_plot2.png")
plt.show()
```
