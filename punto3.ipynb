```python
import numpy as np
import matplotlib.pyplot as plt

z0 = 376.73 # Impedancia caracterstica del vaco
k = (1/2)*z0*(1/(8*np.pi))
theta = np.linspace(0.01, 4*np.pi, 10000)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'},figsize=(10,10))
ax.plot(theta, (np.sin((np.pi)*np.cos(theta))**2)/(np.sin(theta)**2),linewidth=2)
ax.plot(theta, (np.cos((np.pi/2)*np.cos(theta))**2)*(np.cos(theta)**2)/(np.sin(theta)**2), color='r',linewidth=2)

plt.thetagrids([
    0,30,60,90,120,150,210,240,270,300,330, 
])
ax.set_rlabel_position(22.5)  # Move radial labels away from plotted line
ax.grid(True)

#Labels
ax.set_xlabel("ngulo ()")
ax.set_ylabel("Potencia de radiada por unidad de corriente por angulo solido(W/A)")
#Legend
ax.legend(["$\lambda$ = d", "$\lambda$ = 2d"], loc='upper right', bbox_to_anchor=(1.1, 1.1))


ax.set_title("Diagrama de radiacin antena dipolo para $\lambda$ = d y $\lambda$ = 2d")

fig.savefig("Diagrady2d.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

z0 = 376.73 # Impedancia caracterstica del vaco
k = (1/2)*z0*(1/(8*np.pi))
theta = np.linspace(0.01, 4*np.pi, 10000)

fig, ax = plt.subplots(2,1,subplot_kw={'projection': 'polar'},figsize=(8,16))
ax[0].plot(theta, (np.sin((np.pi)*np.cos(theta))**2)/(np.sin(theta)**2),linewidth=2)

##theta grid para 0
ax[0].set_thetagrids([
    0,30,60,90,120,150,210,240,270,300,330,
])
ax[0].set_rlabel_position(22.5)  # Move radial labels away from plotted line

ax[1].plot(theta, (np.cos((np.pi/2)*np.cos(theta))**2)*(np.cos(theta)**2)/(np.sin(theta)**2), color='r',linewidth=2)

plt.thetagrids([
    0,30,60,90,120,150,210,240,270,300,330, 
])
ax[0].set_rlabel_position(22.5)  # Move radial labels away from plotted line
ax[0].grid(True)

#Labels
ax[1].set_xlabel("ngulo ()")
ax[0].set_ylabel("Potencia de radiada por unidad de corriente^2 por angulo solido(W/A)")
ax[1].set_ylabel("Potencia de radiada por unidad de corriente^2 por angulo solido(W/A)")
  
#Legend

ax[0].set_title("Diagrama de radiacin antena dipolo para $\lambda$ = d")
ax[1].set_title("Diagrama de radiacin antena dipolo para $\lambda$ = 2d")

fig.savefig("Diagrad2d.pdf")
plt.show()
```
